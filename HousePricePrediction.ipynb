```python
import pandas as pd
```

```python
HousePrice_pd=pd.read_csv('housing.csv')
HousePrice_pd.head()
```

```python
HousePrice_pd['ocean_proximity'].value_counts()
```

```python
HousePrice_pd_shuffled=HousePrice_pd.sample(n=len(HousePrice_pd),random_state=1)
HousePrice_pd_shuffled
```

```python
pd.get_dummies(HousePrice_pd_shuffled['ocean_proximity']).head()
```

```python
HousePrice_pd_shuffled.drop('ocean_proximity',axis=1).head()
```

```python
HousePrice_pd_final=pd.concat([HousePrice_pd_shuffled.drop('ocean_proximity',axis=1),pd.get_dummies(HousePrice_pd_shuffled['ocean_proximity'])],axis=1)
```

```python
HousePrice_pd_final
```

```python
HousePrice_pd_final=HousePrice_pd_final[['longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population','households','median_income','<1H OCEAN','INLAND','ISLAND','NEAR BAY','NEAR OCEAN','median_house_value']]
HousePrice_pd_final
```

```python
HousePrice_pd_final=HousePrice_pd_final.dropna()
len(HousePrice_pd_final)
```

```python
train_pd,test_pd,val_pd=HousePrice_pd_final[:17000],HousePrice_pd_final[17000:19623],HousePrice_pd_final[19623:]
len(train_pd),len(test_pd),len(val_pd)
```

```python
X_train,y_train=train_pd.to_numpy()[:,:-2],train_pd.to_numpy()[:,-2]
X_val,y_val=val_pd.to_numpy()[:,:-2],val_pd.to_numpy()[:,-2]
X_test,y_test=test_pd.to_numpy()[:,:-2],test_pd.to_numpy()[:,-1]
```

```python
X_train.shape,y_train.shape,X_val.shape,y_val.shape,X_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
scaler=StandardScaler().fit(X_train[:,:5])
def preprocessor(X):
    A=np.copy(X)
    A[:,:5]=scaler.transform(A[:,:5])
    return A
X_train,X_val,X_test=preprocessor(X_train),preprocessor(X_val),preprocessor(X_test)
X_train.shape,X_val.shape,X_test.shape
```

```python
pd.DataFrame(X_train)[0].hist()
```

```python
pd.DataFrame(X_train)[1].hist()
```

```python
#mean squared error
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression
lm=LinearRegression().fit(X_train,y_train)
mse(lm.predict(X_train),y_train,squared=False),mse(lm.predict(X_val),y_val,squared=False)
```

```python
#mean squared error in knn
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=9).fit(X_train,y_train)
mse(knn.predict(X_train),y_train,squared=False),mse(knn.predict(X_val),y_val,squared=False)
```

```python
#mean squared error in random forest
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(max_depth=15).fit(X_train,y_train)
mse(rfr.predict(X_train),y_train,squared=False),mse(rfr.predict(X_val),y_val,squared=False)
```

```python
#mean squared error in gradient boosting
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor(n_estimators=250).fit(X_train,y_train)
mse(gbr.predict(X_train),y_train,squared=False),mse(gbr.predict(X_val),y_val,squared=False)
```
